## DATA PREPARATION (VERİ HAZIRLAMA)

### Modelin eğitilme aşamasından önce verisetimizi bu aşamalara hazırlarız.

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
number_cat = 10
```

```python
y_train
```

```python
import keras
y_train = keras.utils.to_categorical(y_train, number_cat)
```

```python
y_train
```

```python
y_test = keras.utils.to_categorical(y_test, number_cat)
```

```python
y_test
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train
```

```python
X_train.shape
```

```python
Input_shape = X_train.shape[1:]
```

```python
Input_shape
```

##  TRAIN THE MODEL (MODELİN EĞİTİLMESİ)

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, AveragePooling2D, Dense, Flatten, Dropout
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
```

```python
cnn_model = Sequential()
cnn_model.add(Conv2D(filters = 64, kernel_size = (3,3), activation = 'relu', input_shape = Input_shape))
cnn_model.add(Conv2D(filters = 64, kernel_size = (3,3), activation = 'relu'))
cnn_model.add(MaxPooling2D(2,2))
cnn_model.add(Dropout(0.4))


cnn_model.add(Conv2D(filters = 128, kernel_size = (3,3), activation = 'relu'))
cnn_model.add(Conv2D(filters = 128, kernel_size = (3,3), activation = 'relu'))
cnn_model.add(MaxPooling2D(2,2))
cnn_model.add(Dropout(0.4))

cnn_model.add(Flatten())

cnn_model.add(Dense(units = 1024, activation = 'relu'))

cnn_model.add(Dense(units = 1024, activation = 'relu'))

cnn_model.add(Dense(units = 10, activation = 'softmax'))
```

```python
cnn_model.compile(loss = 'categorical_crossentropy', optimizer = keras.optimizers.rmsprop(lr = 0.001), metrics = ['accuracy'])
```

```python
history = cnn_model.fit(X_train, y_train, batch_size = 32, epochs = 1, shuffle = True)
```

## EVALUATE THE MODEL (MODELİN DEĞERLENDİRİLMESİ)

```python
evaluation = cnn_model.evaluate(X_test, y_test)
print('Test Accuracy: {}'.format(evaluation[1]))
```

```python
predicted_classes = cnn_model.predict_classes(X_test) 
predicted_classes
```

## IMPROVING THE MODEL WITH DATA AUGMENTATION (VERİ ARTIŞTIRMA(VERİ BÜYÜTME) İLE MODELİ GELİŞTİRME)

- Image Augmentation is the process of artificially increasing the variations of the images in the datasets by flipping, enlarging, rotating the original images. (Görüntü Büyütme, orijinal görüntüleri çevirerek, büyüterek, döndürerek veri kümelerindeki görüntülerin varyasyonlarını yapay olarak artırma işlemidir.)

- Augmentations also include shifting and changing the brightness of the images.(Büyütmeler ayrıca görüntülerin parlaklığını değiştirmeyi ve değiştirmeyi de içerir.)

## DATA AUGMENTATION FOR THE CIFAR-10 DATASET (CIFAR-10 VERİ SETİ İÇİN VERİ ARTIŞI(VERİ BÜYÜTME))

```python
import keras
from keras.datasets import cifar10
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
X_train.shape
```

```python
n = 8 
X_train_sample = X_train[:n]
```

```python
X_train_sample.shape
```

```python
from keras.preprocessing.image import ImageDataGenerator

# dataget_train = ImageDataGenerator(rotation_range = 90)
# dataget_train = ImageDataGenerator(vertical_flip=True)
# dataget_train = ImageDataGenerator(height_shift_range=0.5)
dataget_train = ImageDataGenerator(brightness_range=(1,3))


dataget_train.fit(X_train_sample)
```

```python
from scipy.misc import toimage

fig = plt.figure(figsize = (20,2))
for x_batch in dataget_train.flow(X_train_sample, batch_size = n):
     for i in range(0,n):
            ax = fig.add_subplot(1, n, i+1)
            ax.imshow(toimage(x_batch[i]))
     fig.suptitle('Augmented images (rotated 90 degrees)')
     plt.show()
     break;
```

## MODEL TRAINING USING AUGEMENTED DATASET (ARTTIRILMIŞ(BÜYÜTÜLMÜŞ) VERİ SETİNİ KULLANARAK MODEL EĞİTİM)

```python
from keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
                            rotation_range = 90,
                            width_shift_range = 0.1,
                            horizontal_flip = True,
                            vertical_flip = True
                             )
```

```python
datagen.fit(X_train)
```

```python
cnn_model.fit_generator(datagen.flow(X_train, y_train, batch_size = 32), epochs = 2)
```

```python
score = cnn_model.evaluate(X_test, y_test)
print('Test accuracy', score[1])
```
